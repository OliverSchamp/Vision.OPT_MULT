```python
import json

json_path = "/home/oliver/Oliver.Mono/projects/Vision.OPT_MULT/data/cropped_data/annotations_all.json"

with open(json_path, 'r') as file:
    annotations_json = json.load(file)

for annotation in annotations_json["annotations"]:
    annotation["iscrowd"] = 0
    annotation["segmentation"] = []

with open(json_path, 'w') as outfile:
    outfile.write(json.dumps(annotations_json))
```
